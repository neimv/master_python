Orientado a objetos

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # los ultimos dos valores son para el tamanio, los primero la pos
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('gray')

fig.show()
```

```python
img, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x, np.cos(x), 'b')
axes[0,1].plot(y, np.sin(x), 'y')
axes[0,2].plot(y, np.tan(x), 'r')
axes[0,3].plot(y, np.cos(x) ** 2)
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x, np.cos(x), 'b')
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x) ** 2)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(x, y, label="$sin(x)$")
axes[0].set_title('Relacion X y Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y, x)
axes[1].set_title('Relacion Y y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x, y, label='sin(x)')
plt.title('este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right', bbox_to_anchor=(0.5, 0.2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A', 'B', 'C'], y=[1, 2, 3])
plt.show()
```

```python
sns.set(style='dark', palette='dark', font='Ubuntu', font_scale=1)
sns.barplot(x=['A', 'B', 'C'], y=[1, 2, 3])
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='dark', alpha=.25)
plt.show()
```

Distribuciones

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist',multiple='stack')
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color="0", marker='<')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, split=True, palette='pastel')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='swarm', col='time') # Define un grafico
plt.show()
```

Relation

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
plt.figure(figsize=(8, 8))
markers = {"Lunch": "D", "Dinner": "s"}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter', col='time')
plt.show()
```

joinplot/pairplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='scatter')
plt.show()
```
